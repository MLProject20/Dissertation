### Percent New Residential Renovation

```python
for row in Housing:
    if row.startswith('NewResReno'):
        Housing['PercNewResReno'+str(row[-4:])] = (Housing['NewResReno'+str(row[-4:])]/Housing['HousingUnits'+str(row[-4:])])*100
```

```python
print(list(Housing.columns))
```

### Percent New Residential Demolition

```python
for row in Housing:
    if row.startswith('NewResDems'):
        Housing['PercNewResDems'+str(row[-4:])] = (Housing['NewResDems'+str(row[-4:])]/Housing['HousingUnits'+str(row[-4:])])*100
```

### Percent New Residential Construction

```python
for row in Housing:
    if row.startswith('NewResCons'):
        Housing['PercNewResCons'+str(row[-4:])] = (Housing['NewResCons'+str(row[-4:])]/Housing['HousingUnits'+str(row[-4:])])*100
```

```python
Housing.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
for row in Housing:
    if row not in ('NPA','Unnamed: 0', 'Unnamed: 0.1') and not row.startswith('Housing')and not row.startswith('ALLViolations'):
        Housing[row[:-4] +'_Z' + row[-4:]] = stats.zscore(Housing[row], nan_policy='omit')
```

```python
year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

for row in year:

    try:

        Housing['HousingRaw'+str(row)] = Housing['NewResReno_Z'+ str(row)] + Housing['NewResDems_Z'+str(row)] + Housing['NewResCons_Z'+str(row)]
        Housing['Social'+str(row)] = Housing['MHIncome_Z'+str(row)] + Housing['MGRent_Z'+str(row)] + Housing['PercentOwnerOccupied_Z'+str(row)]
        Housing['HousingRawPer'+str(row)] = Housing['PercNewResReno_Z'+ str(row)] + Housing['PercNewResDems_Z'+str(row)] + Housing['PercNewResCons_Z'+str(row)]
        Housing['Demographics'+str(row) ] = Housing['PercentWhite_Z'+str(row)] + Housing['MSAge_Z'+str(row)] + Housing['PercentAdultsWithAtLeastBachelors_Z'+str(row)]
        
        
        
    except:
        Housing['Demographics'+str(row) ] = Housing['PercentWhite_Z'+str(row)] + Housing['MSAge_Z'+str(row)] + Housing['PercentAdultsWithAtLeastBachelors_Z2012']
```
